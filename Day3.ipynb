```python
path = ""

with open("Day3.txt") as file:
    for i in file.readlines():
        path += i
        
print(path)
```

```python
santa_grid = []
santa_x = 0
santa_y = 0

robo_grid = []
robo_x = 0
robo_y = 0

path = iter(path)

for i in path:
    print(f"{santa_x},{santa_y}")
    
    santa_grid.append(f"{santa_x},{santa_y}")
    robo_grid.append(f"{robo_x},{robo_y}")
    
    if i == "^":
        santa_x += 1
    elif i == ">":
        santa_y += 1
    elif i == "v":
        santa_x -= 1
    elif i == "<":
        santa_y -= 1
        
    i = next(path)
    
    if i == "^":
        robo_x += 1
    elif i == ">":
        robo_y += 1
    elif i == "v":
        robo_x -= 1
    elif i == "<":
        robo_y -= 1
        
```

```python
grid = robo_grid + santa_grid
```
